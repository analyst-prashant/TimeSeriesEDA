Simple Moving Average

Cumulative Moving Average

Exponential Weighted Moving Average(EWMA)

Moving Average Formula -> Auto Correlation Plot(ACF) (MA)

Autoregressive model -> Partial Auto Correlation Plot(PACF) (AR)

AR + MA = ARIMA


```python
import pandas as pd
import pandas_datareader as pdr
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
```

```python
df_tesla.head()
```

### Simple Moving Average

```python
# Simple Moving Average
df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Open: 10DaysRolling'] = df_tesla['Open'].rolling(window=10).mean()
```

```python
df_tesla.head(10)
```

```python
df_tesla['Open: 10DaysRolling'] = df_tesla['Open'].rolling(window=10, min_periods=2).mean()
# min_periods -> It will show the values from the start
```

```python
df_tesla.head(10)
```

```python
df_tesla['Open: 10DaysRolling'].plot(figsize=(15,6))
```

```python
df_tesla[['Open', 'Open: 10DaysRolling']].plot(figsize=(15,6))
```

```python
df_tesla[['Open', 'Open: 10DaysRolling']].plot(xlim= ['2020-01-01', '2021-01-01'], ylim=(50,800) , figsize=(15,6))
```

```python
df_tesla['Open: 30DaysRolling'] = df_tesla['Open'].rolling(window=30, min_periods=1).mean()
```

```python
df_tesla['Open: 50DaysRolling'] = df_tesla['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df_tesla[['Open', 'Open: 10DaysRolling', 'Open: 30DaysRolling', 
          'Open: 50DaysRolling' ]].plot(xlim= ['2020-01-01', '2021-01-01'], ylim=(50,800) , figsize=(15,6))
```

### Expanding -> CMA(Cumulative Moving Average)

```python
df_tesla['Open:ExpandingMean'] = df_tesla['Open'].expanding().mean()
```

```python
df_tesla.head()
```

```python
df_tesla['Open:ExpandingMax'] = df_tesla['Open'].expanding().max()
```

```python
df_tesla.head()
```

```python
df_tesla['Open:ExpandingMean'].plot(figsize=(12,4))
```

```python
df_tesla[['Open', 'Open:ExpandingMean']].plot()
```

### Exponential Moving Average(EMA)

```python
## EMA Tesla Shares
# Assume smoothening factor = 0.1
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha = 0.1, adjust=False).mean()
```

```python
df_tesla[['Open', 'EMA_0.1' ]].plot(xlim= ['2020-01-01', '2021-01-01'], ylim=(50,800) , figsize=(15,6))
```

```python
# Assume smoothening factor = 0.1
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(alpha = 0.3, adjust=False).mean()
```

```python
df_tesla[['Open', 'EMA_0.1', 'EMA_0.3' ]].plot(xlim= ['2020-01-01', '2021-01-01'], ylim=(50,800) , figsize=(15,6))
```

### Exponential Weighted Moving Average(EWMA)

```python
## EMA Tesla Shares
# Assume smoothening factor = 0.1
df_tesla['EMA_5Days'] = df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open', 'EMA_5Days' ]].plot(figsize=(15,6))
```

```python
## EMA Tesla Shares
# Assume smoothening factor = 0.1
df_tesla['EMA_10Days'] = df_tesla['Open'].ewm(span=10).mean()
```

```python
df_tesla[['Open', 'EMA_5Days', 'EMA_10Days' ]].plot(xlim= ['2020-01-01', '2021-01-01'], ylim=(50,900) , figsize=(15,6))
```
